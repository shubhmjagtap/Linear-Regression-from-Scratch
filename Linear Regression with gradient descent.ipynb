```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
my_data = pd.read_csv('home.txt',names=["size","bedroom","price"])
```

```python
my_data.head()
```

```python
#we need to normalize the features using mean normalization
my_data = (my_data - my_data.mean())/my_data.std()
my_data.head()
```

```python
#setting the matrixes
X = my_data.iloc[:,0:2]
ones = np.ones([X.shape[0],1])

X = np.concatenate((ones,X),axis=1)

y = my_data.iloc[:,2:3].values #.values converts it from pandas.core.frame.DataFrame to numpy.ndarray
theta = np.zeros([1,3])# 1 row and 3 col

#set hyper parameters
alpha = 0.01
iters = 1000
```

```python
#computecost
def computeCost(X,y,theta):
    tobesummed = np.power(((X @ theta.T)-y),2)
    return np.sum(tobesummed)/(2 * len(X))

print(y)
```

```python
#gradient descent

def gradientDescent(X,y,theta,iters,alpha):
    print('X',X)
    print('----------------')
    print('theta.T',theta.T)
    print('----------------')
    print('(X @ theta.T - y)',(X @ theta.T - y))
    print('np.sum(X * (X @ theta.T - y)',np.sum(X * (X @ theta.T - y)))
    cost = np.zeros(iters)
    for i in range(iters):
        theta = theta - (alpha/len(X)) * np.sum(X * (X @ theta.T - y), axis=0)
        cost[i] = computeCost(X, y, theta)
    
    return theta,cost

#running the gd and cost function
g,cost = gradientDescent(X,y,theta,iters,alpha)
print(g)

finalCost = computeCost(X,y,g)
print(finalCost)
```
